## WAFFLE CHART

```python
df_1 = pd.read_csv("/Users/claraferrerrodriguez/UOC/Visualitzacio de dades/exports_spain.csv")
```

```python
df_1 = df_1[['reporterISO', 'partnerISO', 'cifvalue']]

world_total = df_1.loc[df_1["partnerISO"] == "World", "cifvalue"].values[0]
france_value = df_1.loc[df_1["partnerISO"] == "France", "cifvalue"].values[0]
rest_value = world_total - france_value

waffle_df = pd.DataFrame({
    "partner": ["France", "Rest of World"],
    "value": [france_value, rest_value]
})


waffle_df["percentage"] = waffle_df["value"] / waffle_df["value"].sum() * 100


total_squares = 100
squares_per_category = np.round(waffle_df["percentage"]).astype(int)

waffle = []
for i, count in enumerate(squares_per_category):
    waffle.extend([waffle_df["partner"][i]] * count)

#matriu
waffle_matrix = np.array(waffle).reshape((10,10))


fig, ax = plt.subplots(figsize=(6,6))
colors = plt.cm.Set3(np.linspace(0,1,len(waffle_df)))

for i in range(10):
    for j in range(10):
        idx = list(waffle_df["partner"]).index(waffle_matrix[i,j])
        ax.add_patch(
            plt.Rectangle((j, 9-i), 1, 1, facecolor=colors[idx], edgecolor='white'))

ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.set_title("Spain 2024 Exports: France vs Rest of World")


ax.legend(
    handles=[plt.Rectangle((0,0),1,1,color=colors[i]) for i in range(len(waffle_df))],
    labels=waffle_df["partner"].tolist(),  # ← convertir Series a list
    loc="lower center",
    bbox_to_anchor=(0.5,-0.1),
    ncol=len(waffle_df)
)

plt.show()
```

## AREA CHART

```python
df_2 = pd.read_csv("/Users/claraferrerrodriguez/UOC/Visualitzacio de dades/trade_flow.csv")

df_2 = df_2[['refPeriodId','cifvalue']]

df_2['year'] = pd.to_numeric(df_2['refPeriodId'], errors='coerce')
df_2['cifvalue_billions'] = df_2['cifvalue'] / 1e9


df_2 = df_2.sort_values('year')

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
sns.lineplot(data=df_2, x='year', y='cifvalue_billions', marker='o', color='skyblue')
plt.fill_between(df_2['year'], df_2['cifvalue_billions'], color='skyblue', alpha=0.4)

plt.title("Area Plot: CIF Value per Year")
plt.xlabel("Year")
plt.ylabel("CIF Value (Billion USD)")
plt.xticks(df_2['year'].unique(), rotation=45)
plt.tight_layout()
plt.show()
```
